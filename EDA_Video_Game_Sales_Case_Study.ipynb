```python
#Year wise top selling and least selling games
top_selling = dict()
least_selling = dict()
```

```python
years = list(df['Year'].unique())
years = sorted(years)
years
```

```python
for i in years:
    top_selling[i] = list()
    
for i in years:
    least_selling[i]= list()
```

```python
for i in years:
    x = df.loc[df['Year'] == i]
    res = x['Name'].loc[x['Global_Sales']==x['Global_Sales'].max()]
    top_selling[i].insert(0,res.values[0])
```

```python
x
```

```python
x['Global_Sales'].max()
```

```python
res
```

```python
top_selling
```

```python
for i in years:
    x = df.loc[df['Year'] == i]
    res = x['Name'].loc[x['Global_Sales']==x['Global_Sales'].min()]
    least_selling[i].insert(0,res.values[0])
```
